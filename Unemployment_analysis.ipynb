```python
#month wise unemplloyment rate with respect to estimated employed and estimated Labour participation rate
import plotly.graph_objects as go

df_EEE = df.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']].mean()
df_EEE = pd.DataFrame(df_EEE).reset_index()
month = df_EEE.month
Unemployment_rate = df_EEE['Estimated Unemployment Rate']
Labour_participation_rate = df_EEE['Estimated Labour Participation Rate']

fig = go.Figure()
fig.add_trace(go.Bar(x=month, y=Unemployment_rate, name='Unemployment_Rate'))
fig.add_trace(go.Bar(x=month, y=Labour_participation_rate, name='Labour Participation Rate'))

fig.update_layout(
    title='Unemployment rate and Labour Participation Rate',
    xaxis={'categoryorder': 'array', 'categoryarray': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']},
    yaxis=dict(title='Rate (%)') # Add y-axis label for rates
)
fig.show()
```

```python
df_EEE11 = df_11.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']].mean()
df_EEE11 = pd.DataFrame(df_EEE11).reset_index()
month = df_EEE11.month
Unemployment_rate = df_EEE11['Estimated Unemployment Rate']
Labour_participation_rate = df_EEE11['Estimated Labour Participation Rate']

fig = go.Figure()
fig.add_trace(go.Bar(x=month, y=Unemployment_rate, name='Unemployment_Rate'))
fig.add_trace(go.Bar(x=month, y=Labour_participation_rate, name='Labour Participation Rate'))

fig.update_layout(
    title='Unemployment rate and Labour Participation Rate',
    xaxis={'categoryorder': 'array', 'categoryarray': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']},
    yaxis=dict(title='Rate (%)') # Add y-axis label for rates
)
fig.show()
```
